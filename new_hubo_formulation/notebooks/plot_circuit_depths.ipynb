```python
import pickle
from matplotlib import pyplot as plt
import numpy as np
```

```python
def read_data(res: dict[str, dict]):
    filenames = ('test_N2_W2', 'test_N7_W2', 'trivial', 'test_N8_W2', 'test_N3_W4', 'test_N8_W3', 'test_N7_W4', 'test_N4_W5')
    # filenames = list(res.keys())
    # res[filename]['default']['circuit'].num_qubits
    qubits = [res[filename]['default']['circuit'].num_qubits for filename in filenames]
    

    default_depths, default_counts = [], []
    rzz_depths, rzz_counts = [], []

    for filename in filenames:
        default_count, default_depth = res[filename]['default']['count'], res[filename]['default']['depth']
        default_depths.append(default_depth)
        default_counts.append(default_count)
        
        rzz_count, rzz_depth = res[filename]['rzz']['count'], res[filename]['rzz']['depth']
        rzz_depths.append(rzz_depth)
        rzz_counts.append(rzz_count)
                
    # qubits, filenames, default_depths, default_counts, rzz_depths, rzz_counts = zip(*sorted(zip(
    #     qubits, filenames, default_depths, default_counts, rzz_depths, rzz_counts
    # )))

        
    default_counts = np.array(default_counts)
    rzz_counts = np.array(rzz_counts)

    default_depths = np.array(default_depths)
    rzz_depths = np.array(rzz_depths)
    return qubits, default_depths, default_counts, rzz_depths, rzz_counts, filenames
```
