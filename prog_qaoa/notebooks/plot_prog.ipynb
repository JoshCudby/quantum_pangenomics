```python
import numpy as np
```

```python
n = 6
T = 3
counts = {
    '000000000': 2,
    '101011111': 4
}
```

```python
rng = np.random.default_rng()
random_samples = rng.integers(1, n+1, (sum(counts.values()), T), endpoint=True)

random_samples = [
    [np.binary_repr(x, int(np.ceil(np.log2(n+2)))) for x in random_sample] for random_sample in random_samples
]
random_samples = [
    ''.join(random_sample) for random_sample in random_samples
]
random_samples
```
