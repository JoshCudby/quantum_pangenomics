```python
from dwave.optimization import Model
import networkx as nx
```

```python
t_max = 4
N = 3
model = Model()
variables = model.integer((t_max,), 0, N + 1)
```

```python
g = nx.DiGraph()
g.add_nodes_from([
    (0, {"weight": 1}),
    (1, {"weight": 1}),
    (2, {"weight": 1}),
    (3),
    # (4, {"weight": 1}),
    # (5, {"weight": 1}),
    # (6, {"weight": 1}),
    # (7),
])
g.add_edges_from([
    (0, 1),
    (0, 2),
    (0, 3),
    # (0, 7),
    (1, 2),
    (1, 3),
    # (1, 7),
    (2, 3),
    (3, 3)
    # (2, 7),
    # (3, 4),
    # (3, 5),
    # (3, 6),
    # (3, 7),
    # (4, 7),
    # (5, 6),
    # (5, 7),
    # (6, 4),
    # (6, 7),
    # (7, 7),
])
```

```python
for t in range(t_max - 1):
    model.add_constraint((variables[t], variables[t + 1]) in g.edges)
```
