```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
solver='mqlib'
```

```python
with open(f'/nfs/users/nfs_j/jc59/quantumwork/pangenome/out/2.eval.{solver}.coverages3.txt', 'r') as f:
    coverages_3 = f.readlines()
coverages_3 = [float(x[0:-1]) for x in coverages_3]  
```

```python
with open(f'/nfs/users/nfs_j/jc59/quantumwork/pangenome/out/2.eval.{solver}.coverages30.txt', 'r') as f:
    coverages_30 = f.readlines()
coverages_30 = [float(x[0:-1]) for x in coverages_30]  
```

```python
fig, axs = plt.subplots(1,1,figsize=(8, 5))

bin_width = 0.25
bins = np.arange(np.min([coverages_3, coverages_30]),
                    np.max([coverages_3, coverages_30]) + bin_width, bin_width)

axs.hist(coverages_3, weights=np.ones_like(coverages_3), bins=100, label='Time limit 3', density=False, alpha=0.7)
axs.hist(coverages_30, weights=-np.ones_like(coverages_30), bins=100, label='Time limit 30', density=False, alpha=0.7)
ylims = axs.get_ylim()

axs.axhline(0, color="k")

axs.legend()
axs.set_xlabel("(Coverage% + Usage%) / 2")
axs.set_ylabel("Sample count")

axs.set_title(f'Histogram of {solver} runs')

fig.savefig(f'/nfs/users/nfs_j/jc59/quantumwork/pangenome/out/{solver}.coverage.histogram.png')
```

```python
with open(f'/nfs/users/nfs_j/jc59/quantumwork/pangenome/out/2.eval.{solver}.breaks3.txt', 'r') as f:
    breaks_3 = f.readlines()
breaks_3 = [int(x[0:-1]) for x in breaks_3]  
```
