```python
import numpy as np
import sympy as sp
from sympy.solvers import solve
```

```python
n = 2
O = sp.symbols(f'O(0:2)(0:{2*n})')
O = np.array(list(O))
O = O.reshape((2, 2 * n))
```

```python
eqns = []
for i in range(2 * n):
    for j in range(i+1, 2 * n):
        eqns.append(O[0, i] * O[1, j] - O[0, j] * O[1, i] - (1 if (i, j) == (0, 1) else 0))
```

```python
sol = solve(eqns, O[1, 2])
```
